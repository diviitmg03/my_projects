```python
import pandas as pd

df = pd.read_csv('predictions.csv')
df.head()
```

```python
predictions = df['pred'].tolist()
references = df['ref'].tolist()
```

```python
from rouge_score import rouge_scorer

scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
scores = [scorer.score(ref, pred) for ref, pred in zip(references, predictions)]

# Calculate average ROUGE scores
avg_rouge1 = sum([score['rouge1'].fmeasure for score in scores]) / len(scores)
avg_rouge2 = sum([score['rouge2'].fmeasure for score in scores]) / len(scores)
avg_rougeL = sum([score['rougeL'].fmeasure for score in scores]) / len(scores)

print(f"Avg ROUGE-1: {avg_rouge1}")
print(f"Avg ROUGE-2: {avg_rouge2}")
print(f"Avg ROUGE-L: {avg_rougeL}")
```

```python
import sacrebleu

bleu = sacrebleu.corpus_bleu(predictions, [references])
print(f"BLEU score: {bleu.score}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Model names and their corresponding perplexity scores
models = ['T5', 'Flan-T5', 'BART', 'GODEL']
perplexities = [2.10, 6.20, 3.75, 3.40]
target = 2.5  # This is the benchmark or target perplexity

# Create a figure and axes
fig, ax = plt.subplots(figsize=(8, 5))

# Create horizontal bars for each model's perplexity
for i, (model, perp) in enumerate(zip(models, perplexities)):
    # Add a light gray background bar to represent the target
    ax.barh(i, 10, color='lightgray', height=0.4)  # Adjust the width and height of the bar
    # Add the actual perplexity value as a darker bar
    ax.barh(i, perp, color='blue', height=0.4)
    # Add a red line indicating the target perplexity
    ax.vlines(target, -0.5, len(models) - 0.5, colors='red', linestyles='dashed')
    # Add text for the actual perplexity value
    ax.text(perp + 0.1, i, f'{perp:.2f}', va='center')

# Customize the axes
ax.set_yticks(np.arange(len(models)))
ax.set_yticklabels(models)
ax.set_xlim(0, 10)
ax.set_xlabel('Perplexity Score')
ax.set_title('Perplexity Comparison for Different Models')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Model names and perplexity scores
models = ['T5', 'Flan-T5', 'BART', 'GODEL']
perplexities = [2.10, 6.20, 3.75, 3.40]

# Number of variables (perplexities) and the angle of each axis
num_vars = len(models)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Repeat the first value to close the circular chart
perplexities += perplexities[:1]
angles += angles[:1]

# Create a radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# Draw the radar chart
ax.fill(angles, perplexities, color='blue', alpha=0.25)
ax.plot(angles, perplexities, color='blue', linewidth=2)

# Add labels for each axis
ax.set_yticklabels([])  # Hide the y-tick labels (optional)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(models)

# Set the title and labels
plt.title('Perplexity Comparison for Different Models', size=15, color='black', y=1.1)

# Show the radar chart
plt.tight_layout()
plt.show()
```
